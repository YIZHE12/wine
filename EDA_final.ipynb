```r
par(mfrow = c(3, 4))  # Set up a 2 x 2 plotting space

# Create the loop.vector (all the columns)
loop.vector <- 1:12

for (i in loop.vector) { # Loop over loop.vector

  # store data in column.i as x
  x <- Data_red[,i]
  
  # Plot histogram of x
  hist(x,
       main = paste(colnames(Data_red)[i]),
       xlab = "Scores")
   
}
```
